```python
from bs4 import BeautifulSoup

import urllib
import urllib.request
import re
import pandas as pd
```

```python
with urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States') as response:
    html = response.read()
```

```python
soup = BeautifulSoup(html, "lxml")
type(soup)
```

```python
th= soup.find_all('th')
states = []
for tag_th in th:
            a = tag_th.find_all('a')
            for tag_a in a:
                states.append(tag_a.get_text())
print("['"+"', '".join(states)+"']")
```

```python
states = [item for item in states if len(item) > 3 ]
states = states[9:59]
print("['"+"', '".join(states)+"']", len(states))
```

```python
td= soup.find_all('td')
population = []
for tag_td in td:
            div = tag_td.find_all('div')
            for tag_div in div:
                population.append(tag_div.get_text())
print("['"+"', '".join(population[:101])+"']")
```

```python
population = population[:101]
population = [item.replace(',', '') for item in population if len(item) > 4]
print("['"+"', '".join(population)+"']")
```

```python
PDstates = pd.DataFrame(states, columns=['State'])
PDpopulation = pd.DataFrame(population, columns=['Population'])
table = pd.concat([PDstates, PDpopulation], axis=1)
```
